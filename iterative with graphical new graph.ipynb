```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_node('D')
G.add_node('E')
G.add_node('F')

G.add_edge ('B', 'C')
G.add_edge ('A', 'D')
G.add_edge ('A', 'E')
G.add_edge ('A', 'F')
G.add_edge ('B', 'F')
G.add_edge ('C', 'E')
G.add_edge ('D','F')

plt.figure()
nx.draw(G, with_labels=True)
plt.show()
```
